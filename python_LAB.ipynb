7. given a signal value=[ 10,25,40,55,70]
normalise them between 0 and 1 using the formula  x'= x-min(x)/max(x)-min(x)

```python

val=[10,25,40,55,70]
max_x = max(val)
min_x = min(val)
for v in val:
  print((v-min_x)/(max_x-min_x))
```

9. write a program to generate a digital logic truth table.

```python
A=[0,1]
B=[0,1]
print('"|" AND "| " OR "|" XOR "|"')
for a in A:
  for b in B:
    print((a & b)  ,(a | b),  (a^b))
```
