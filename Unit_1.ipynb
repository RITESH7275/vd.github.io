# "is" keyword vs "==" comparison.
1. **is keyword:-**
Checks identity → whether two variables refer to the same object in memory.  
Syntax: a is b  
True only if both a and b point to the exact same object (same memory location).  

**2. == operator:-**
Checks equality → whether the values of two objects are the same.  
Syntax: a == b  
True if the data/content inside them is equal, even if stored in different objects.  


**"is" keyword**

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)  # True  (**both refer to same object**)
print(a is c)  # False (**different objects with same values**)
print(a is not c)
```

**"==" operator**

```python
a = [1, 2, 3]
c = [1, 2, 3]
print(a == c)  # True  (values are same)
print(a is c)  # False (different objects)
```
